```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

# Interpolando la función $Seno(x)$ con Lagrange

```python
def funcion_analitica(x):
    return np.sin(x)
```

```python
def lagrange(n,xl,yl):
    """Guarda los polinomios"""
    polinomio = 0 #POLINOMIO SOLUCION
    polinomios = [] #GUARDO CADA POLINOMIO
    for i in range(n): #sumas
        lagranges = 1.0
        for j in range(n): #producto
            if i != j:
                lagranges = lagranges*(symbols('x')-xl[j])  / (xl[i] - xl[j])
        polinomios.append(lagranges*yl[i])
        polinomio = polinomio +lagranges*yl[i]
        
    return polinomio, polinomios
            
xl = np.linspace(0.,2.,21)
yl = funcion_analitica(xl)

print xl

n = len(xl)
poli, polinomios = lagrange(n,xl,yl)
```

```python
puntos = np.linspace(.0,2.,101)
fpuntos = funcion_analitica(puntos)
```

```python
from sympy import *
from sympy.abc import x

polys = []

plt.figure(figsize=(5,5))

#Principal polinomio
expr = poli
f = lambdify(x, expr)
yp = f(puntos)

plt.plot(puntos,yp, color = "purple", label= "Solucion Interpolacion")
plt.plot(puntos,funcion_analitica(puntos), color = "green", label= "Solucion Real")
plt.plot(puntos,funcion_analitica(puntos)-yp, color = "black", label = "error")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1))
```

Para este ejemplo podemos ver que el error es casi 0 y la interpolación es un éxito. Casi está superpuesta en la curva originalde la función $seno(x)$

# Tabla de Valores y Error 

```python
import pandas as pd

resultados = pd.DataFrame({"Sol Analitica":funcion_analitica(puntos) ,"Interpolacion": yp,
             "DIFF":(funcion_analitica(puntos)-yp)})
resultados.tail()
```

# Tarea : Fenómeno de Runge

Dentro del intervalo x $\in$ [-1,1] calcule valores de la función 

$f(x) = \frac{1}{1+25x^{2}}$ 

Utilizando interpolación de Lagrange calcule h = 200

$x_{i} = -1 + h(i-1) $

$y_{i} = f(x_{i}) + \varepsilon* r$

$h = \frac{2}{n-1}$


```python
def funcion_analitica(x):
    return 1./(1.+(25.*(x**2)))
```

```python
#7 nodos equidistantes
n = 7
xl = np.array([-1 +(i-1)*(2./(n-1)) for i in range(1,n+1)])

r = np.random.rand(len(xl))
error = r*0.001
y = funcion_analitica(xl)
yl = y+error

#Interpolacion
n = len(xl)
poli, __ = lagrange(n,xl,yl)

expr = poli
f = lambdify(x, expr)

xd = np.linspace(-1,1,500)
ya = funcion_analitica(xd)
yi = f(xd)

plt.scatter(xd,ya,c="r", label = "Funcion Runge", alpha=.5)
plt.scatter(xd,yi,c="b", label="Interpolacion 7 nodos equidistantes", alpha=.5)
plt.scatter(xl, funcion_analitica(xl), c="y")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1))
```

**Comentarios** : Como podemos ver los nodos que se usaron para hacer la interpolación (en este caso 7) no tienen un error. Pero el resto de la curva solución tiene gran error hacia los extremos y además oscila.

```python
#16 nodos equidistantes
n = 10
xl = np.array([-1 +(i-1)*(2./(n-1)) for i in range(1,n+1)])

r = np.random.rand(len(xl))
error = r*0.001
y = funcion_analitica(xl)
yl = y+error

#Interpolacion
n = len(xl)
poli, __ = lagrange(n,xl,yl)

expr = poli
f = lambdify(x, expr)

xd = np.linspace(-1,1,500)
ya = funcion_analitica(xd)
yi = f(xd)

plt.scatter(xd,ya,c="r", label = "Funcion Runge", alpha=.5)
plt.scatter(xd,yi,c="b", label="Interpolacion 10 nodos equidistantes", alpha=.5)
plt.scatter(xl, funcion_analitica(xl), c="y")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1))
```
